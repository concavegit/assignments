# Creating a Caltech Dataset Wrapper Using ImageFolder

This is a streamlined version of the notebook we sent earlier on how to create a dataset wrapper for the Caltech 256 Dataset.  Please see the rest of that notebook for how to take this dataset and train a model.

```
import gdown
gdown.download('https://drive.google.com/uc?authuser=0&id=1GMtKmoH2Yu9yY6jx4Mm5fZ1sLkGV9i_c&export=download',
               'caltech256.zip',
               quiet=False)
!unzip -qq caltech256.zip
```

```
from torchvision.datasets import ImageFolder
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

def show_image(img_tensor):
    # need to reorder the tensor dimensions to work properly with imshow
    plt.imshow(img_tensor.transpose(0,2).transpose(0,1))
    plt.axis('off')
    plt.show()

cal_tech = ImageFolder(root='./256_ObjectCategories', transform=transforms.Compose([transforms.CenterCrop((200,200)),
                                                    transforms.Resize((100,100)),
                                                    transforms.ToTensor()]))

im, target = cal_tech[10000]
print(cal_tech.classes[target])
show_image(im)
print(im.shape)
```
